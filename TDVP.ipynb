```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
totalt = 4000
dt = 0.05
t = np.linspace(1, totalt, totalt)*dt

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))

ees = np.load("figureMain/ees_Main1test.npy", allow_pickle = True)[:,1]
ccs = np.load("figureMain/ccs_Main1test.npy", allow_pickle = True)
pts = 2*np.load("figureMain/Pts_Main1test.npy", allow_pickle = True)
mzs = np.load("figureMain/mzs_Main1test.npy", allow_pickle = True)

ax1.set(title = "$\Delta = $ -0.8~0.5, m = 0.2, D=32", ylabel ='Entanglement entropy', xlim = (0, 42), xlabel = 'real time t')
ax1.plot(t, ees, color = 'orange')
ax2.plot(t, ccs)
ax2.set(ylabel ='Chiral condensate', ylim = (0.065, 0.18), xlim = (0, 42), xlabel = 'real time t')
ax3.plot(t, pts, color = 'r')
ax3.set(ylabel ='$-\ln P(t)$', xlim = (0, 42), xlabel = 'real time t')
plt.show()

fig2, ax = plt.subplots()
ax.plot(t, mzs,color = 'c')
ax.set(xlabel='real time t', ylabel='$m_z$', xlim = (0, 42), ylim=(-0.01,0.01))
# ax.grid()
plt.show()
```

```python
Ds = [40, 50, 80]
totalt = 280
dt = 0.05
t = np.linspace(1, totalt, totalt)*dt

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))

for D in Ds:
    ees = np.load("figureA7/eesD%d.npy"%D, allow_pickle = True)
    ax1.set(title="$\Delta = 0.5$, m = 0 ~ 0.5", ylabel = 'Entanglement entropy', xlabel = 'real time t', xlim = (0, 14) )
    ax1.plot(t, ees, label = 'D=%d'%D)
    ax1.legend()

for D in Ds:
    ccs = np.abs(np.load("figureA7/ccsD%d.npy"%D, allow_pickle = True))
    ax2.set(ylabel = 'Chiral condensate', xlabel = 'real time t', xlim = (0, 14) )
    ax2.plot(t, ccs, label = 'D=%d'%D)
    ax2.legend()

for D in Ds:
    pts = 2*np.load("figureA7/ptsD%d.npy"%D, allow_pickle = True)
    ax3.set(ylabel = '$-\ln P(t)$', xlabel = 'real time t', xlim = (0, 14) )
    ax3.plot(t, pts, label = 'D=%d'%D)
    ax3.legend()
```

```python
Ds = [20, 32, 80]
t = np.linspace(1, 280, 280)*0.05
t2 = np.linspace(1, 2000, 2000)*0.01
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))

for D in Ds:
    ees = np.load("figureMain/ees2D%d.npy"%D, allow_pickle = True)
    ax1.set(title="$\Delta = -0.8$ ~ 0.5, m = 0.2", ylabel = 'Entanglement entropy', xlabel = 'real time t', xlim = (0, 25) )
    if D==80:
        ax1.plot(t2, ees, label = 'D=%d'%D)       
    else:
        ax1.plot(t, ees, label = 'D=%d'%D)
    ax1.legend()

for D in Ds:
    ccs = np.abs(np.load("figureMain/ccs2D%d.npy"%D, allow_pickle = True))
    ax2.set(ylabel = 'Chiral condensate', xlabel = 'real time t', xlim = (0, 25) )
    
    if D==80:
        ax2.plot(t2, ccs, label = 'D=%d'%D)       
    else:
        ax2.plot(t, ccs, label = 'D=%d'%D)
        
    ax2.legend()

for D in Ds:
    pts = 2*np.load("figureMain/Pts2D%d.npy"%D, allow_pickle = True)
    ax3.set(ylabel = '$-\ln P(t)$', xlabel = 'real time t', xlim = (0, 25) )
    
    if D==80:
        ax3.plot(t2, pts, label = 'D=%d'%D)       
    else:
        ax3.plot(t, pts, label = 'D=%d'%D)
        
    ax3.legend()
```
