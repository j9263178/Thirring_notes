## Quench: $\Delta = $ -0.8~0.5, m = 0.2

```python
img = mpimg.imread('Main1.png')
plt.figure(figsize = (8,8))
imgplot = plt.imshow(img, interpolation='none')
plt.axis('off')
plt.show()

t = [np.linspace(1, 4000, 4000)*0.05, np.linspace(1, 800, 800)*0.05]
D = [32, 64]
c = ['burlywood', 'deeppink']
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16,4))
fig2, ax = plt.subplots()

for i in range(2):
    
    datas = np.load("figureMain/datas_Main1_D%d.npy"%D[i], allow_pickle = True)
    ax1.set(title = "$\Delta = $ -0.8~0.5, m = 0.2, D=%d"%D[i], ylabel ='Entanglement entropy', xlim = (0, 42), xlabel = 'real time t')
    ax1.plot(t[i], datas[0], label ='D = %d'%D[i], color = c[i])
    ax2.set(ylabel ='Chiral condensate', ylim = (0.065, 0.18), xlim = (0, 42), xlabel = 'real time t')
    ax2.plot(t[i], datas[1], label ='D = %d'%D[i], color = c[i])
    ax3.set(ylabel ='$-\ln P(t)$', xlim = (0, 42), xlabel = 'real time t')
    ax3.plot(t[i], 2*datas[2], label ='D = %d'%D[i], color = c[i])
    ax.set(xlabel='real time t', ylabel='$m_z$', xlim = (0, 42), ylim=(-0.01,0.01))
    ax.plot(t[i], datas[3], label ='D = %d'%D[i], color = c[i])

ax1.legend()
ax2.legend()
ax3.legend()
ax.legend()
```

## Traverse field Ising model
### $ H = \sum_{r \in \mathbb{Z}} -\sigma_x ^{r} \sigma_x^{r+1} - h_z \sigma_z^r$

```python

img = mpimg.imread('tfim.png')
plt.figure(figsize = (12,12))
plt.title("Result from iTEBD with h = 10 $\to$ h = 3 [G. Vidal 2007]")
imgplot = plt.imshow(img, interpolation='none')
plt.axis('off')
plt.show()

dt = 0.01
ts = [np.linspace(1, 1000, 1000)*dt ,np.linspace(1, 600, 600)*dt , np.linspace(1, 500, 500)*dt]
Ds = [40, 20,10]
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16,4))
ax1.set(title = "TFIM, h = 10 ~  h = 3", ylabel ='Entanglement entropy', xlabel = 'real time t')
exact = np.load("mzexact.npy")

for i in range(3):
    datas = np.load("ising/datas_isingD%d.npy"%Ds[i], allow_pickle = True)
    ax1.plot(ts[i], datas[0], label = 'D = %d'%Ds[i])
    ax2.set(xlabel='real time t', ylabel='$m_z$', xlim = (0,  10))
    ax2.plot(ts[i], np.abs(datas[3])/2, label = 'D = %d'%Ds[i])
ax1.legend()
ax2.legend()

fig2, ax = plt.subplots(figsize=(10,4))
ax.set(title = 'megnetization $m_z$', xlabel='real time t', ylabel='$m_z$', xlim = (0,  10), ylim = (0.475, 0.495))
ax.plot(ts[0], 0.5*(exact-exact[0])+0.49874764643055725, label='Exact', color = 'violet',linewidth=4.5)
for i in range(3):
    datas = np.load("ising/datas_isingD%d.npy"%Ds[i], allow_pickle = True)
    ax.plot(ts[i], np.abs(datas[3])/2, label = 'D = %d'%Ds[i])
ax.legend()

fig3, ax0 = plt.subplots(figsize=(10,4))
ax0.set(title = 'Errors in megnetization  $(m_z - \hat{m_z})/\hat{m_z}$', xlabel='real time t', ylabel='relative errors for $m_z$', xlim = (0,  10), ylim = (-0.0006, 0.0006))
exact_ = 0.5*(exact-exact[0])+0.49874764643055725
exacts = [exact_, exact_[:600], exact_[:500]]
# ax0.plot(ts[0], label='Exact', color = 'violet',linewidth=4.5)
for i in range(3):
    datas = np.load("ising/datas_isingD%d.npy"%Ds[i], allow_pickle = True)
    ax0.plot(ts[i], (np.abs(datas[3])/2 - exacts[i])/exacts[i], label = 'D = %d'%Ds[i])
ax0.legend()
```

## Old TDVP results for Thirring (before optimization)
